```python
import numpy as np
from time import time
from matplotlib.pyplot import plot
N=10**5

def I(f,a,b):
    global N
    A=0
    for k in range(N):
        dx= (b-a)/N
        dA= dx/2 * (f(a+k*dx) + f(a+(k+1)*dx))
        A+=dA
    return A

def P(f,a,x):
    return I(f,a,x)
    

def f(x):
    return x**2 - x + 3

def g(x):
    return np.exp(x) - 2*np.cos(x)

def h(x):
    return 2 - 3*x**3 + np.exp(-x)


def F(x):
    return x**3/3 - x**2/2 + 3*x

def G(x):
    return np.exp(x) - 2*np.sin(x)

def H(x):
    return 2*x - 3/4*x**4 - np.exp(-x)
```

```python
a=float(input('Inserisci a: '))
b=float(input('Inserisci b: '))

t=time()
I1=I(f,a,b)
I2=F(b)-F(a)
print('\nIntegrale stimato di f:',I1)
print('Integrale esatto di f: ',I2)
print('Errore percentuale:', 100*(I1-I2)/I2,'%.')

I1=I(g,a,b)
I2=G(b)-G(a)
print('\nIntegrale stimato di g:',I1)
print('Integrale esatto di g: ',I2)
print('Errore percentuale:', 100*(I1-I2)/I2,'%.')

I1=I(h,a,b)
I2=H(b)-H(a)
print('\nIntegrale stimato di h:',I1)
print('Integrale esatto di h: ',I2)
print('Errore percentuale:', 100*(I1-I2)/I2,'%.')
t=time()-t

print('\nTempo di esecuzione:',round(t,5),'s.')
```

```python
X=np.linspace(a,b,N)
t=time()
print('Seguono i grafici di f(in rosso), g(in blu), h(in verde).\n')
z=plot(X,f(X),'red')
z=plot(X,g(X),'blue')
z=plot(X,h(X),'green')
t=time()-t
print('Tempo di esecuzione:',round(t,5),'s.')
```
