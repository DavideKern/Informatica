```python
import numpy as np
import matplotlib.pyplot as plt
M=10**4

def f(x):
    return 4/np.pi**4 * (x**2 - np.pi**2) * (x**2 - (np.pi/2)**2)

def fk(x,k):
    return f(x) * np.cos(k*x)

def I(k):
    A=0
    dx=2*np.pi/M
    for i in range(M):
        dA= dx/2 * (fk(-np.pi+i*dx,k) + fk(-np.pi+(i+1)*dx,k))
        A+=dA
    return A

C=[]
for k in range(10):
    C.append( round(I(k)/np.pi,7) )

for N in range(0,10):
    def S(x,N):
        S=C[0]/2
        k=1
        while k<=N:
            S+= C[k]*np.cos(k*x)
            k+=1
        return S

X=np.linspace(-np.pi, np.pi, M)
Y=np.zeros((10,M))
for N in range(10):
    Y[N]=S(X,N)

def d(x,N):
    return (f(x) - S(x,N))**2

def dist(N):
    A=0
    dx = 2*np.pi/M
    for i in range(M):
        dA= dx/2 * (d(-np.pi + i*dx, N) + d(-np.pi + (i+1)*dx, N))
        A+=dA
    return A

D=[]
for k in range(10):
    D.append( round(dist(k),7) )
```
