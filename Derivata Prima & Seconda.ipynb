```python
from time import time
import numpy as np
from matplotlib.pyplot import plot

h=10**(-5)
k=10**(-8)
N=5000

def f(x):
    return np.log(x)

def f1(x):
    global h
    return (f(x+h) - f(x))/h

def f2(x):
    global k
    return (f1(x+k) - f1(x))/k

a=0.001
b=10

X=np.linspace(a,b,N)
```

```python
print('Grafico f(x):')
p=plot(X,f(X),'red')
```

```python
print("Grafico f'(x):")
p=plot(X,f1(X),'orange')
```

```python
print("Grafico f''(x):")
p=plot(X,f2(X),'gold')
```
